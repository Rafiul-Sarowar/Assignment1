```python
# to see unique data types

df.dtypes.unique()
```

```python
#to check which variables has integer data type

df1= df.select_dtypes(include= "int64")
df1_col_list= df1.columns.tolist()

len(df1_col_list)
```

```python
#to check variables with object type data

df2= df.select_dtypes(include="float64")
df2_col_list= df2.columns.tolist()
len(df2_col_list)
```

```python
# variables with float value

df3= df.select_dtypes(include="O")
df3_col_list= df3.columns.tolist()
len(df3_col_list)
```

```python
#to check duplicate values

duplicates= df.duplicated()
df[duplicates]
```

```python
# to check columns with missing values

check= df.isnull().sum() #creates a series named 'missing_values' 
missing_cols= check[check>0]
missing_cols.index
len(missing_cols.index)
```

```python
#checking unique values of columns with null values

#for 'int64' dtype

m1=df1.isnull().sum()
m1
series1= m1[m1>0]
len(series1)
```

```python
m2= df2.isnull().sum()
m2
series2= m2[m2>0]
# series2
# len(series2)
```

```python
m3= df3.isnull().sum()
m3
series3= m3[m3>0]
series3
len(series3)
```

```python
#checking unique values of columns with null values

#for 'float' dtype

missingcols2= series2.index
# for col in missingcols2:
#     print("Unique value for", col, "is", df2[col].unique())

#replace missing values with avg values
for col in missingcols2:
    df2[col]=df2[col].fillna(df2[col].mean())

df2
```

```python
# df2.isnull().sum()
```

```python
#checking unique values of columns with null values

#for 'object' dtype

missingcols3= series3.index
# for col in missingcols3:
#     print("Unique value for", col, "is", df3[col].unique())

for col in missingcols3:
    df3[col]=df3[col].fillna("None")

df3
```

```python
# df3.isnull().sum()
```

```python
#outlier check for a column (sale price)

Q1= df["SalePrice"].quantile(0.25)
Q3= df["SalePrice"].quantile(0.75)

IQR= Q3-Q1
lower_bound= Q1- 1.5*IQR
Upper_bound= Q3+ 1.5*IQR

df4= df[(df["SalePrice"]<lower_bound) | (df["SalePrice"]> Upper_bound)]
df4["SalePrice"]
```
